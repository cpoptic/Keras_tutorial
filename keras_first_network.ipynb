```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
# fix random seed for reproducibility
np.random.seed(7)
```

```python
# load pima indians dataset
dataset = np.loadtxt("..\Jason_Brownlee\Datasets\pima-indians-diabetes.data.csv", delimiter=",")
```

```python
dataset
```

```python
type(dataset)
```

```python
# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# Define Model
# create a sequential model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy', metrics.categorical_accuracy, 'mse', 'mae', 'mape', 'cosine'])
# There are several metrics out of the obx with Theano, so I'm just trying a few of them here.
```

```python
# Fit the model
model.fit(X, Y, epochs=150, batch_size=10)
```

```python
# evaluate the model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
model.metrics_names
```
